## Preprocessing

```python
df['Loan_Type'].head()
```

```python
def count_strings(val):
    if(pd.isna(val)):
        return 0
    else:
        return len(val.split(','))

df['Loan_Count'] = df['Loan_Type'].apply(count_strings)
df['Loan_Count']
df_test['Loan_Count'] = df_test['Loan_Type'].apply(count_strings)
df_test['Loan_Count'].head()
```

```python
df['Credit_History_Age'].head()
```

```python
def convert_to_months(val):
    if pd.isna(val):
        return 0
    else:
        years, months = int(val.split(" ")[0]), int (val.split(" ")[3])
    return years * 12 + months
```

```python
df['Credit_History_Age'] = df['Credit_History_Age'].apply(convert_to_months)
df['Credit_History_Age'].head()
df_test['Credit_History_Age'] = df_test['Credit_History_Age'].apply(convert_to_months)
df_test['Credit_History_Age'].head()
```

```python
numeric_cols = ['Age', 'Income_Annual', 'Base_Salary_PerMonth','Total_Bank_Accounts', 'Total_Credit_Cards', 'Rate_Of_Interest',
                'Total_Current_Loans', 'Delay_from_due_date','Total_Delayed_Payments', 'Credit_Limit','Total_Credit_Enquiries',
                'Current_Debt_Outstanding','Ratio_Credit_Utilization', 'Per_Month_EMI','Monthly_Investment', 'Monthly_Balance',
                'Loan_Count']

for col in numeric_cols:
    if df[col].dtype == 'object':
        df[col] = df[col].str.replace(r'[^-.0-9]', '', regex=True)
    df[col] = pd.to_numeric(df[col], errors='coerce')

for col in numeric_cols:
    df[col] = df.groupby('Customer_ID')[col].transform(lambda x: x.fillna(x.median()))
    df[col].fillna(df[col].median(), inplace=True)
```

```python
numeric_cols = ['Age', 'Income_Annual', 'Base_Salary_PerMonth','Total_Bank_Accounts', 'Total_Credit_Cards', 'Rate_Of_Interest',
                'Total_Current_Loans', 'Delay_from_due_date','Total_Delayed_Payments', 'Credit_Limit','Total_Credit_Enquiries',
                'Current_Debt_Outstanding','Ratio_Credit_Utilization', 'Per_Month_EMI','Monthly_Investment', 'Monthly_Balance',
                'Loan_Count']

for col in numeric_cols:
    if df_test[col].dtype == 'object':
        df_test[col] = df_test[col].str.replace(r'[^-.0-9]', '', regex=True)
    df_test[col] = pd.to_numeric(df_test[col], errors='coerce')

for col in numeric_cols:
    df_test[col] = df_test.groupby('Customer_ID')[col].transform(lambda x: x.fillna(x.median()))
    df_test[col].fillna(df_test[col].median(), inplace=True)
```

```python
x1 = 'Credit_History_Age'
overall_mode = df[x1].mode()[0]

# Impute NaN values within each Customer_ID group with the mode of that group
df[x1] = df.groupby('Customer_ID')[x1].transform(lambda x: x.fillna(x.mode()[0] if not x.mode().empty else overall_mode))
df_test[x1] = df_test.groupby('Customer_ID')[x1].transform(lambda x: x.fillna(x.mode()[0] if not x.mode().empty else overall_mode))

# Fill any remaining NaN values with the overall mode
df[x1].fillna(overall_mode, inplace=True)
df_test[x1].fillna(overall_mode, inplace=True)

# Print the number of NaNs left in the column
print(df[x1].isna().sum())
print(df_test[x1].isna().sum())
```

```python
issue_cols = ['Credit_Mix', 'Payment_of_Min_Amount', 'Profession', 'Payment_Behaviour']
for col in issue_cols:
    print(f'Unique values of {col}: {df[col].unique()}')
    print(f'Unique values of {col}: {df_test[col].unique()}')
```

```python
# Dictionary of categorical columns with issues
issue_categorical_dict = {
    "Credit_Mix": "_",
    "Payment_Behaviour": "!@9#%8",
    "Payment_of_Min_Amount": "NM",
    "Profession": "_______"}

# Impute categorical columns by Customer_ID group mode, falling back to overall mode
for col, problem_value in issue_categorical_dict.items():
    if col in df.columns:
        # Define the overall mode, excluding problematic values
        overall_mode = df.loc[df[col] != problem_value, col].mode().get(0, 'Unknown')

        # Replace the problematic value with group mode; fallback to overall mode if needed
        df[col] = df.groupby("Customer_ID")[col].transform(
            lambda x: x.replace(problem_value, x[x != problem_value].mode().get(0, overall_mode)))

        # Print value counts to check imputation results
        print(f"Value counts for {col}:\n", df[col].value_counts())
```

```python
# Dictionary of categorical columns with issues

issue_categorical_dict = {
    "Credit_Mix": "_",
    "Payment_Behaviour": "!@9#%8",
    "Payment_of_Min_Amount": "NM",
    "Profession": "_______"}

# Impute categorical columns by Customer_ID group mode, falling back to overall mode
for col, problem_value in issue_categorical_dict.items():
    if col in df_test.columns:
        # Define the overall mode, excluding problematic values
        overall_mode = df_test.loc[df_test[col] != problem_value, col].mode().get(0, 'Unknown')
        # Replace the problematic value with group mode; fallback to overall mode if needed
        df_test[col] = df_test.groupby("Customer_ID")[col].transform(
            lambda x: x.replace(problem_value, x[x != problem_value].mode().get(0, overall_mode)))

        # Print value counts to check imputation results
        print(f"Value counts for {col}:\n", df_test[col].value_counts())
```

```python
for col in categorical_columns:
    df[col] = df.groupby('Customer_ID')[col].transform(lambda x: x.fillna(x.mode()[0] if not x.mode().empty else np.nan))
    df[col].fillna(df[col].mode()[0], inplace=True)
```

```python
categorical_columns = ['Credit_Mix', 'Payment_Behaviour', 'Payment_of_Min_Amount', 'Month', 'Profession']
categorical_columns = [col for col in categorical_columns if col in df.columns]
for col in categorical_columns:
    df_test[col] = df_test.groupby('Customer_ID')[col].transform(lambda x: x.fillna(x.mode()[0] if not x.mode().empty else np.nan))
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
# Apply label encoding to 'Profession' column

le = LabelEncoder()
df['Profession'] = le.fit_transform(df['Profession'])
df_test['Profession'] = le.transform(df_test['Profession'])

# Apply label encoding to other categorical columns
for col in categorical_columns:
    if col != 'Profession':  # Skip 'Profession' as it's already encoded
        df[col] = le.fit_transform(df[col])
        df_test[col] = le.transform(df_test[col])

# Drop unnecessary columns (IDs, names, or any unique identifiers)
columns_to_drop = ['ID', 'Name', 'Number', 'Loan_Type', 'Customer_ID']
df = df.drop(columns=[col for col in columns_to_drop if col in df.columns])
df_test = df_test.drop(columns=[col for col in columns_to_drop if col in df_test.columns])
```

```python
credit_score_mapping = {'Poor': 0, 'Standard': 1, 'Good': 2}
df['Credit_Score'] = df['Credit_Score'].map(credit_score_mapping)
```
